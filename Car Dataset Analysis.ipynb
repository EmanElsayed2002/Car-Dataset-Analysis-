```python
cars = df.select_dtypes(include=['float64' , 'int64'])
cars
```

```python
cars.mean()
```

```python
cars = cars.fillna(cars.mean())
```

```python
cars.isna().sum()
```

```python
cars["num-of-doors"] = df["num-of-doors"]
```

```python
cars["num-of-cylinders"] = df["num-of-cylinders"]
```

```python
cars["fuel-type"] = df["fuel-type"]
```

```python
cars["make"] = df["make"]
```

```python
cars
```

```python
cars.isna().sum()
```

```python
cars["num-of-cylinders"].value_counts()
```

```python
cars.columns
```

```python
cars["num-of-cylinders"] = cars["num-of-cylinders"].fillna(cars["num-of-cylinders"].mode())
```

```python
cars["num-of-doors"] = cars["num-of-doors"].fillna(cars["num-of-doors"].mode())
```

```python
cars.info()
```

```python
cars#.value_counts()
```

```python
cars["num-of-doors"].value_counts()
```

```python
cars["num-of-doors"] = cars["num-of-doors"].fillna(cars["num-of-doors"].mode()[0])
```

```python
cars["num-of-doors"] = cars["num-of-doors"].replace({"two": 2 , "four": 4})
```

```python
cars["num-of-cylinders"] = cars["num-of-cylinders"].replace({"four" : 4 , "six": 6 , "five": 5 , "eight": 8 , "two": 2 , "three": 3 , "twelve": 12})
```

```python
cars.info()
```

```python
cars["num-of-doors"].value_counts()
```

```python
corr = cars.select_dtypes(include = ["int64"  , "float64"]).corr()
corr
```

```python
plt.figure(figsize=(20 , 15))
sns.heatmap(corr, annot = True) 
plt.show()
```
